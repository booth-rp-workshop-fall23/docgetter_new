# Archive

```python
# This block details the addition of old slide collections completed by Sonali and Esfandiar in Q3 2023

# all old files from the slide process were download manually, and as such, are uploaded separately into a special folder at:
# /project/FactSet/Doc_Retrieval_API/factset_slides_API/output/OLD_SLIDE_DOWNLOADS

# these slides do not correspond to slide_IDs so can only be retrieved by firm_name string

# their file_names are not uniquely named so the unique identifer had to be manually created. 
# this is done by adding the firm_name (latter part of the folder_name) as a prefix to the 'file_name' column in trans_ref
# note that ACTUAL file names have not changed -- only how their names are represented on trans_ref

```

```python
import os
import re

old_name_adds = []

#  collect names and addresses of all files in old_slide_downlaod such that they can be added to docgetter
for energy_util in os.listdir('../OLD_SLIDE_DOWNLOAD/'):
    if energy_util.startswith('.'):
        continue
    # get energy/utility folder prefix
    pre_prefix = energy_util
    for folder_name in os.listdir(f'../OLD_SLIDE_DOWNLOAD/{pre_prefix}'):
        if folder_name.startswith('.'):
            continue
        # get firm name -- remove number + underscore
        firm_name = folder_name.split('_')[-1]
        # standardize firm name by removing whitespace, adding underscore etc.
        firm_name = re.sub('\s', '_', firm_name)
        for f_name in os.listdir(f'../OLD_SLIDE_DOWNLOAD/{pre_prefix}/{folder_name}'):
            if f_name.startswith('.'):
                continue
            #file_name = firm_name + '_' + f_name
            address = f'./OLD_SLIDE_DOWNLOAD/{pre_prefix}/{folder_name}'
            old_name_adds.append([f_name, address])


df = pd.DataFrame(old_name_adds, columns=['file_name', 'address'])
print(df.shape)
df.sample(5)
```

```python
df[~df.file_name.duplicated(keep=False)].sort_values('file_name')
```

```python
# concat these rows to trans_ref
new_ref = pd.concat([new_ref, df])
new_ref
```
